```python
import requests
import pandas as pd

#On récupère les information de connexion
APP_ID = 'beb2d3b8'
APP_KEY = '9bcf67b2b619db30ab3dd2d56118b4d8'

# URL de l'API Edamam
url = "https://api.edamam.com/api/food-database/v2/parser"

# Paramètres de requête
params = {
    'app_id': APP_ID,
    'app_key': APP_KEY, 
    'ingr': 'Champagne'
}

# On envoi la requête à l'API
response = requests.get(url, params=params)

# On vérifie si on a une réponse
if response.status_code == 200:
    data = response.json()

    # On prépare une liste pour stocker les produits
    products = []

    # Boucle sur les 10 premiers produits
    for i, product in enumerate(data['hints'][:10]):
        food = product['food']
        product_data = {
            'foodId': food['foodId'],
            'label': food['label'],
            'category': food.get('category', 'N/A'),
            'foodContentsLabel': food.get('foodContentsLabel', 'N/A'),
            'image': food.get('image', 'N/A')
        }
        products.append(product_data)

    # Création d'un DataFrame
    df = pd.DataFrame(products)

    # Sauvegarde du DataFrame en fichier CSV
    df.to_csv('products.csv', index=False)

    print("Extraction réussie. Les données ont été sauvegardées dans 'products.csv'.")
else:
    print("Erreur lors de la requête : ", response.status_code)
```
